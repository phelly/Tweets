```python
tweets = []


def text_query_to_csv(text_query,count):
    try:
        #Pulling individual tweets from query
        for tweet in api.search(q=text_query, count=count):
        
        #Adding to list that contains all tweets
         tweets.append((tweet.created_at,tweet.id,tweet.text,tweet.source,tweet.favorite_count,tweet.retweet_count,tweet.username))
        
        # creation of the dataframe from tweets list
        tweets_df = pd.DataFrame(tweets, columns = ['Datetime', 'Tweet_Id', 'Text', 'Source', 'Favorite Count', 'Retweets'])
        
        #converting from dataframe to csv
        tweets_df.to_csv('{}-tweets.csv'. format(text_query))
        
        
    except BaseException as e:
        print('failed on_status,', str(e))
        time.sleep(3)
```
